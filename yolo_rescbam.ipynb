```python
from ultralytics import YOLO

model = YOLO("/content/drive/MyDrive/project/yolov8-rescbam/runs/detect/yolov8m-rescbam-finetune/weights/last.pt")
model.train(
    data="/content/drive/MyDrive/project/realdata2/data.yaml",
    epochs=50,                 # Sá»‘ epoch huáº¥n luyá»‡n thÃªm
    batch=8,
    imgsz=640,
    name="yolov8m-rescbam-finetune",  # TÃªn phiÃªn train má»›i
    device=0,
    amp=False,                # Táº¯t AMP náº¿u tá»«ng gÃ¢y lá»—i
    resume=True,             # KhÃ´ng dÃ¹ng resume mÃ  dÃ¹ng fine-tune
    pretrained=True,          # Váº«n giá»¯ pretrained True
    lr0=1e-4,                 # ğŸ’¡ Giáº£m learning rate Ä‘á»ƒ fine-tune mÆ°á»£t hÆ¡n
    warmup_epochs=0,          # KhÃ´ng cáº§n warmup láº¡i
    close_mosaic=10,          # Giá»¯ nhÆ° cÅ© náº¿u mosaic Ä‘ang Ä‘Æ°á»£c dÃ¹ng
    patience=30               # Early stopping sau 30 epoch khÃ´ng cáº£i thiá»‡n
)
```

```python
model = YOLO("/content/drive/MyDrive/project/yolov8-rescbam/runs/detect/yolov8m-rescbam-finetune/weights/best.pt")
metrics = model.val() # Assign the validation results to the 'metrics' variable
print("mAP@0.5:", metrics.box.map50)
print("Precision:", metrics.box.mp) # Use the mp() method for mean precision
print("Recall:", metrics.box.mr)    # Use the mr() method for mean recall
print("mAP@0.5:0.95:", metrics.box.map)
print("F1-Score:", metrics.box.f1)  # Use the f1() method for mean F1-score
```
