```python
from ultralytics import YOLO
import torch

model = YOLO("/content/drive/MyDrive/project/FCE-YOLOv8-main/runs/detect/tea_disease_yolov8m_globalcontext/weights/best.pt/")
metrics = model.val()
print("mAP@0.5:", metrics.box.map50)
print("Precision:", metrics.box.mp) # Use the mp() method for mean precision
print("Recall:", metrics.box.mr)    # Use the mr() method for mean recall
print("mAP@0.5:0.95:", metrics.box.map)
print("F1-Score:", metrics.box.f1)  # Use the f1() method for mean F1-score
```

```python
from ultralytics import YOLO

model = YOLO("/content/drive/MyDrive/project/FCE-YOLOv8-main/runs/detect/tea_disease_yolov8m_globalcontext/weights/best.pt")
model.train(
    data="/content/drive/MyDrive/project/realdata2/data.yaml",
    epochs=50,                 # Số epoch huấn luyện thêm
    batch=8,
    imgsz=640,
    name="yolov8m-GC-finetune",  # Tên phiên train mới
    device=0,
    amp=False,                # Tắt AMP nếu từng gây lỗi
    resume=False,             # Không dùng resume mà dùng fine-tune
    pretrained=True,          # Vẫn giữ pretrained True
    lr0=1e-4,                 # 💡 Giảm learning rate để fine-tune mượt hơn
    warmup_epochs=0,          # Không cần warmup lại
    close_mosaic=10,          # Giữ như cũ nếu mosaic đang được dùng
    patience=30               # Early stopping sau 30 epoch không cải thiện
)
```

```python
model = YOLO("/content/drive/MyDrive/project/FCE-YOLOv8-main/runs/detect/yolov8m-GC/weights/best.pt/")
metrics = model.val()
print("mAP@0.5:", metrics.box.map50)
print("Precision:", metrics.box.mp) # Use the mp() method for mean precision
print("Recall:", metrics.box.mr)    # Use the mr() method for mean recall
print("mAP@0.5:0.95:", metrics.box.map)
print("F1-Score:", metrics.box.f1)  # Use the f1() method for mean F1-score
```
